```python
from google.colab import drive
import pandas as pd
import numpy as np

def main():
    # 1. Google Drive 마운트
    print("0. Google Drive 연동을 시작합니다...")
    drive.mount('/content/drive')

    file_path = '/content/drive/MyDrive/ACK2026_Wafer/MIR-WM811K/MIR-WM811K/Python/WM811K_defects.pkl'

    # 2. 데이터 로드
    print("\n1. 데이터를 불러오는 중입니다...")
    try:
        df = pd.read_pickle(file_path)
        print("-> 로드 완료!\n")
    except FileNotFoundError:
        print(f"오류: '{file_path}' 파일을 찾을 수 없습니다.")
        return

    print("=" * 50)
    print(" 📊 웨이퍼 크기 및 dieSize 분포 분석")
    print("=" * 50)

    # ----------------------------------------------------
    # [분석 1] waferMap 행렬 크기(Shape) 분포 확인
    # ----------------------------------------------------
    print("\n[ 1. waferMap 행렬 크기(Shape) 분포 ]")
    # waferMap의 각 행렬에서 shape 속성(예: (26, 26))을 추출하여 문자열로 변환한 뒤 파생 변수 생성
    df['wafer_shape'] = df['waferMap'].apply(lambda x: str(x.shape))

    # 크기별로 개수 집계
    shape_counts = df['wafer_shape'].value_counts()
    print(f"- 고유한 행렬 크기 종류: 총 {len(shape_counts):,}개")
    print("- 상위 15개 행렬 크기 분포 (가장 많은 순):")
    print(shape_counts.head(15)) # 상위 15개만 출력

    print("-" * 50)

    # ----------------------------------------------------
    # [분석 2] dieSize (총 칩의 개수) 분포 확인
    # ----------------------------------------------------
    print("\n[ 2. dieSize (총 칩의 개수) 분포 ]")
    # dieSize 값별로 개수 집계
    dieSize_counts = df['dieSize'].value_counts()
    print(f"- 고유한 dieSize 종류: 총 {len(dieSize_counts):,}개")
    print("- 상위 15개 dieSize 분포 (가장 많은 순):")
    print(dieSize_counts.head(15)) # 상위 15개만 출력


if __name__ == "__main__":
    main()
```
